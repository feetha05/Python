# Merging Two Data Frames

```python
# import modules
import pandas as pd


staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},{'Name': 'Sally', 'Role': 'Course Liason'},
                        {'Name': 'James', 'Role': 'Grader'}])

student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business'},{'Name': 'Mike', 'School': 'Law'},
                        {'Name': 'Sally', 'School': 'Engineering'}])

# set the Name as the index

student_df = student_df.set_index('Name')
staff_df = staff_df.set_index('Name')
```

```python
# Syntax
#inner join = intersection
merge1 = pd.merge(staff_df,student_df, how = 'outer', left_index = True , right_index = True) 
merge1
```

```python
# Syntax
#inner join = intersection
merge2 = pd.merge(staff_df,student_df, how = 'outer', left_index = True , right_index = True) 
merge2
```

```python
# left join to get only all staff
#left join to only get the staff

merge3 = pd.merge(staff_df,student_df, how = 'inner', left_index = True , right_index = True) 
merge3
```

```python
# right join to get only all staff
#right join to only get the staff

merge4 = pd.merge(staff_df,student_df, how = 'right', left_index = True , right_index = True) 
merge4
```

```python
# in the case where more than one key is neccesary for the join , you need to use a list of the intems to right join
# or left join 
mergetwo = (pd.merge(staff_df,student_df, how = 'inner', left_on = ['FirtName','Last Name'], 
                     right_on=['First Name','Last Name']))
```
